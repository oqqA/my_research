```
import plotly.graph_objects as go
import numpy as np

import json
```

```
# Parser file from STM32CubeMonitor

loadData = json.load(open('./data/_.json'))

newData = [[],[],[]]

for i in range(6, len(loadData), 9):
    for j in range(3):
        for v in loadData[i+j]['variabledata']:
            newData[j].append(v['y'])

with open('data/_new.json', 'w') as file:
    json.dump(newData, file)
```

```
import json
data = json.load(open('./data/fixZmagNew.json'))
```

```
t = np.linspace(0, 20, len(data[0]))

fig = go.Figure(data=[go.Scatter3d(
    x=data[0],
    y=data[1],
    z=data[2],
    mode='markers',
    marker=dict(
        size=8,
        color=t,                # set color to an array/list of desired values
        colorscale='Viridis',   # choose a colorscale
        opacity=0.4
    )
)])

# tight layout
fig.update_layout(template='plotly_dark', margin=dict(l=0, r=0, b=0, t=0))
fig.show()
```
